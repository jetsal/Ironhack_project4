### Check out some confidence intervals of mean vaccination rates

```python
#WORLD
from scipy.stats import t

df = len(covid)

t = t.interval(0.99, df)[1]

mean = covid["vacc_100"].mean() 

# Standard error
standard_error_world = covid["vacc_100"].std() / np.sqrt(len(covid["vacc_100"]))


con_int_world = [mean - (t * standard_error_world) , mean + (t * standard_error_world)]
print('With 99 % confidence, the true world mean vaccination rate lies in the inverval:', con_int_world)
```

```python
#WORLD MINUS EU/WESTERN EUROPEAN COUNTRIES
from scipy.stats import t

non_eu = covid[covid["region"]!="WE/EU"]
df = len(non_eu)

t = t.interval(0.99, df)[1]

mean = non_eu["vacc_100"].mean() 

# Standard error
standard_error_non_eu = non_eu["vacc_100"].std() / np.sqrt(len(non_eu["vacc_100"]))


con_int_non_eu = [mean - (t * standard_error_non_eu) , mean + (t * standard_error_non_eu)]
print('With 99 % confidence, the true non_EU-world mean vaccination rate lies in the inverval:', con_int_non_eu)
```

```python
# EU WESTERN EUROPEAN COUNTRIES
from scipy.stats import t
europe = covid[covid["region"]=="WE/EU"]
df = len(europe)

t = t.interval(0.99, df)[1]

mean = europe["vacc_100"].mean() 

# Standard error
standard_error_europe = europe["vacc_100"].std() / np.sqrt(len(europe["vacc_100"]))


con_int_europe = [mean - (t * standard_error_europe) , mean + (t * standard_error_europe)]
print('With 99 % confidence, the true EU mean vaccination rate lies in the inverval:', con_int_europe)
```

### Corruption: hypothesis

H0: The vaccination rate between the two compared countries does not differ.
H1: The vaccination rate between the two compared countries does differ.

H0: There <b>is no</b> significant correlation between corruption index of the compared countries and their vaccination rate <br><br>
H1: There <b>is</b> a significant correlation between corruption index of the compared countries and their vaccination rate



### Corruption: two sided t-test

I am inclined to think corruption ruins a country, but it does not have to negatively affect the vaccination rate.<br>
Quicker access to vaccin for certain groups of people through corruption?<br>

Follow up question: does corruption affect <b>who</b> is vaccinated.

```python
st, p = ttest_ind(hun_pol["CPI"], hun_pol["vacc_100"], equal_var=False)
alpha = 0.025

if p<alpha:
    outcome = "rejected"
else:
    outcome = "accepted"

print("HUNGARY VERSUS POLAND")
print("H0 is:", outcome)
print("statistic:", round(st,4), "p-value:", round(p,5))
```

```python
covid2 = covid[["country", "ISO3", "GDP", "LN_GDP", "vacc_100", "region", "nr_vaccins", "CPI", "GINI", "HCare_GDP"]].dropna(axis=0)
```

### Look at relationship between % of GDP spent on healthcare and vaccination rate

<li>H0: % of GDP spent op healthcare is NOT correlated with vaccination rate</li>
<li>H1: higher % of GDP spent on healtcare results in higher vaccination rate</li>

Note: one tailed testing, if p=< 0.05 Ã¡nd the t-statistic is positive, H0 is rejected.

```python
X = covid2["HCare_GDP"].tolist()
Y = covid2["vacc_100"].tolist()

X = sm.add_constant(X)
result = sm.OLS(Y, X).fit()
print(result.summary())
```

There is a significant correlation (in the right direction) between % of GDP spent on healthcare and vaccination rate. So H0 should be rejected.
But... it would make sense that the wealth of the country plays a role in how much % of the GDP is spent on healthcare. therefore it would be better to control for GDP. This is done in the next cell.

H0: % GDP spent on healthcare or GDP are NOT correlated with vaccination rate
H1: % GDP spent on healthcare and/or GDP are positively correlated with vaccination rate (the higher %GDP spent on healthcare or GDP, the higher the vacc rate).

Note: one sided testing. Only if p=< 0.05 Ã¡nd the t-statistic is positive, H0 is rejected.

```python
X = covid2[["HCare_GDP", "LN_GDP"]]
Y = covid2["vacc_100"]

X = sm.add_constant(X)
result = sm.OLS(Y, X).fit()
print(result.summary())
```

In both cases, H0 is rejected. The explanation of the model hugly improved by adding GDP (from ~17 to 54%)

#### Normalize % spent on healthcare GDP and GDP

```python
mean_HC = covid2["HCare_GDP"].mean()
std_HC = covid2["HCare_GDP"].std()

mean_LN_GDP = covid2["LN_GDP"].mean()
std_LN_GDP = covid2["LN_GDP"].std()

covid2["norm_HC"] = (covid2["HCare_GDP"] - mean_HC) / std_HC
covid2["norm_LN_GDP"] = (covid2["LN_GDP"]- mean_LN_GDP) / std_LN_GDP
```

```python
X = covid2[["norm_HC", "norm_LN_GDP"]]
Y = covid2["vacc_100"]

X = sm.add_constant(X)
result = sm.OLS(Y, X).fit()
print(result.summary())
```
