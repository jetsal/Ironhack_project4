```python
#TRY THE MULTI_CURSOR-CLICKING IN THIS EXAMPLE
fig, ax = plt.subplots(1, 1)

df_rest = 55

alpha_left = 0.025
alpha_left = t.ppf(alpha_left, df=df_rest)
alpha_right = 1 - 0.025
alpha_right = t.ppf(alpha_right, df=df_rest)

x = np.linspace(-5,5, 100)
y=t.pdf(x, df=df_rest)

ax.plot(x,y)

plt.fill_between(x,y, where = x >= alpha_right, color='g')
plt.fill_between(x,y, where = x <=alpha_left, color='g')
ax.plot(x, t.pdf(x=x, df = df_rest), 'k', lw=3);
plt.show()
```

```python
fig, ax = plt.subplots(1, 1)

df_rest = 55

alpha_left = 0.025
alpha_left = t.ppf(alpha_left, df=df_rest)
alpha_right = 1 - 0.025
alpha_right = t.ppf(alpha_right, df=df_rest)

x = np.linspace(-5,5, 100)
y=t.pdf(x, df=df_rest)
plt.plot([4], [0.03], 'v', markersize = 10, lw=4, color='b')
plt.plot([-4], [0.03], 'v', markersize = 10, lw=4, color='b')

ax.plot(x,y)

plt.fill_between(x,y, where = x >= alpha_right, color='g')
plt.fill_between(x,y, where = x <=alpha_left, color='g')
ax.plot(x, t.pdf(x=x, df = df_rest), 'k', lw=3);
plt.show()
```
